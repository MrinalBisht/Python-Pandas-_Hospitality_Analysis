**Data Transformation**

```python
df_agg_bookings.head(5)
```

```python
df_agg_bookings["occ_pct"] = df_agg_bookings['successful_bookings']/df_agg_bookings["capacity"]
```

```python
df_agg_bookings.head(5)
```

```python
df_agg_bookings["occ_pct"] = df_agg_bookings["occ_pct"].apply(lambda x: round(x*100, 2))
df_agg_bookings.head(5)
```

**Data Insights**

**1. What is an average occupany rate in each of the room categories?**

```python
df_agg_bookings.groupby("room_category")["occ_pct"].mean()
```

```python
df_agg_bookings.groupby("room_category")["occ_pct"].mean().round(2) #For rounding up mean values
```

```python
df_rooms
```

```python
df = pd.merge(df_agg_bookings, df_rooms, left_on = "room_category", right_on = "room_id")
df.tail(7)
```

```python
df.groupby("room_class")["occ_pct"].mean().round(2)
```

```python
df.drop('room_id', axis = 1, inplace = True)
df.head(7)
```

**2. Print average occupany per city**

```python
df_hotels.head(5)
```

```python
df = pd.merge(df, df_hotels, on = "property_id")
df.head(7)
```
